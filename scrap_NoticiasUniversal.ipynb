```python
import requests
```

```python
import bs4
```

```python
response = requests.get('https://rpp.pe/')
soup = bs4.BeautifulSoup(response.content,'html.parser')
titles = soup.select('.cont a')
titles
```

```python
titles_labels = [title['href'] for title in titles ]
for i in titles_labels:
    print(i)
```
